```python
import pandas as pd
from sqlalchemy import create_engine
```

### Store CSV into DataFrame

```python
csv_file = "Resources/winemag-data_first150k.csv"
wine_data_df = pd.read_csv(csv_file)
wine_data_df.head()
```

### Create new data with select columns

```python
wine_data_us_df = wine_data_df[wine_data_df.country == "US"]
wine_data_us_df.head()
```

```python
wine_data_loc_df = wine_data_us_df
wine_data_loc_df["Lat"] = ''
wine_data_loc_df["Lng"] = ''

wine_data_loc_df.head()
```

```python
import gmaps
import requests
# !pip install gmaps
```

```python
from gconfig import (gkey)
gmaps.configure(api_key=gkey)

params = {"key": gkey}

for index, row in wine_data_loc_df.iterrows():
    base_url = "https://maps.googleapis.com/maps/api/geocode/json"

    state = row['province'] 
    params['address'] = f"{state}"
    state_lat_lng = requests.get(base_url, params=params)
    state_lat_lng = state_lat_lng.json()

    wine_data_loc_df.loc[index, "Lat"] = state_lat_lng["results"][0]["geometry"]["location"]["lat"]
    wine_data_loc_df.loc[index, "Lng"] = state_lat_lng["results"][0]["geometry"]["location"]["lng"]

wine_data_loc_df.head()
```
